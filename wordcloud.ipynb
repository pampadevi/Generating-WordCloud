```python
# open the file and read it into a variable alice_novel
File = open('indian_menu.txt', 'r').read()
    
print ('File downloaded and saved!')
```

```python
# install wordcloud
!conda install -c conda-forge wordcloud==1.4.1 --yes

# import package and its set of stopwords
from wordcloud import WordCloud, STOPWORDS

print ('Wordcloud is installed and imported!')
```

```python
import numpy as np  # useful for many scientific computing in Python
import pandas as pd # primary data structure library
from PIL import Image # converting images into arrays

%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches # needed for waffle Charts

mpl.style.use('ggplot') # optional: for ggplot-like style

# check for latest version of Matplotlib
print ('Matplotlib version: ', mpl.__version__) # >= 2.0.0


print("Done")
```

```python
stopwords = set(STOPWORDS)
```

```python
# instantiate a word cloud object
Pampa_wc = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stopwords
)

# generate the word cloud
Pampa_wc.generate(File)
```

```python
# display the word cloud
plt.imshow(Pampa_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Lets make the image bigger

```python
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

# display the cloud
plt.imshow(Pampa_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Remove words that we dont like 

```python
stopwords.add('Sauce') # add the words said to stopwords
stopwords.add('Onion')
stopwords.add('lamb')
stopwords.add('cooked')
stopwords.add('sauteed')
#stopwords.add('Sukhi')
#stopwords.add('India')
#stopwords.add('Prerna Singh')
#stopwords.add('Guide')





# re-generate the word cloud
Pampa_wc.generate(File)

# display the cloud
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(Pampa_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

save the image to a png file

```python
Pampa_wc.to_file("menu.png")
```

We can add any arbitrary mask and lay the image on top of it 

```python
# download image
#!wget --quiet https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Images/alice_mask.png
    
# save mask to alice_mask
cv_mask = np.array(Image.open('cv_mask.png'))
    
print('Image downloaded and saved!')
```

```python
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(cv_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# instantiate a word cloud object
menu_wc = WordCloud(background_color='white', max_words=2000, mask=cv_mask, stopwords=stopwords)

# generate the word cloud
menu_wc.generate(File)

# display the word cloud
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(menu_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
